## XOR

```python
import numpy as np

# вагові коефіцієнти для входів
W1 = np.array([[1, -1], [-1, 1]])
W2 = np.array([[1], [1]])
```

```python
def activation(t):
    return 0 if t < 0.5 else 1


def predict(x):
    t1 = x @ W1  # @ - умножение матрицы на вектор
    h1 = list(map(activation, t1))
    t2 = h1 @ W2
    y = activation(t2)
    return y
```

```python
# вхідні параметри 1
x = np.array([0, 0])

probable = predict(x)
print('prediction: ', probable)
classes = ['False', 'True']
print('Predicted class: ', classes[probable])
```
